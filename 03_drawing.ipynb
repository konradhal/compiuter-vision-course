###<a name='a2'></a> Wczytanie obrazu

````
img = cv2.imread('czerwonyNG1.png')
cv2_imshow(img)
````

###<a name='a3'></a> Dodanie prostokąta

````
copy = img.copy()
cv2_imshow(cv2.rectangle(img=copy, pt1=(75, 80), pt2=(180, 200), color=(0, 0, 255), thickness=2))
````

````
copy = img.copy()
cv2_imshow(cv2.rectangle(img=copy, pt1=(40, 50), pt2=(50, 70), color=(255,0, 0), thickness=2))
````

###<a name='a4'></a> Dodanie koła/okręgu

````
copy = img.copy()
cv2_imshow(cv2.circle(img=copy, center=(100, 100), radius=30, color=(255, 0, 0), thickness=-1))
````

````
copy = img.copy()
cv2_imshow(cv2.circle(img=copy, center=(300, 300), radius=20, color=(255, 0, 0), thickness=3))
````

###<a name='a5'></a> Dodanie linii

````
copy = img.copy()
cv2_imshow(cv2.line(img=copy, pt1=(180, 300), pt2=(300, 300), color=(255, 0, 0), thickness=2))
````

###<a name='a6'></a> Dodanie tekstu

````
copy = img.copy()
cv2_imshow(cv2.putText(img=copy, text='ELEMENT NG! WYPAD ', org=(5, 30), 
                       fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1,
                       color=(0, 255, 0), thickness=2))
````
